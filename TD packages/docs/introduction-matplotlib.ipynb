## 1. 2 façons de créer des tracés

Il y a deux façons principales de créer des tracés dans matplotlib.

1. `matplotlib.pyplot.plot()` - Recommandé pour les tracés simples (par exemple x et y).
2. `matplotlib.pyplot.XX` (où XX peut être une des nombreuses méthodes, c'est ce qu'on appelle l'API orientée objet) - Recommandé pour les tracés plus complexes (par exemple [`plt.subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots) pour créer plusieurs tracés sur la même Figure, nous y reviendrons plus tard). 

Ces deux méthodes sont encore souvent créées en construisant à partir de `import matplotlib.pyplot as plt` comme base.

Commençons simplement.

````python
# Traçons des points
plt.plot([1, 2, 3, 4]);
````

````python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
````

Quelques petites choses à propos d'un graphique :
* `x` est l'axe horizontal.
* `y` est l'axe vertical.
* Dans un point de données, `x` vient habituellement en premier, par exemple `(3, 4)` serait `(x=3, y=4)`.
* La même chose se produit dans [`matplotlib.pyplot.plot()`] (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html), `x` vient avant `y`, par exemple `plt.plot(x, y)`.

````python
# Traçons une courbe
plt.plot(x, y);
````

Essayons maintenant d'utiliser la version orientée objet.

Nous allons commencer par créer une figure avec `plt.figure()`.

Puis nous ajouterons un axe avec [`add_subplot`](https://matplotlib.org/stable/api/figure_api.html#matplotlib.figure.Figure.add_subplot). 

````python
# Création d'un graphique avec la version orientée objet
fig = plt.figure() # créer une figure
ax = fig.add_subplot() # ajouter un axe 
plt.show()
````

Une note sur la terminologie :
* Une **`Figure`** (par exemple `fig = plt.figure()`) est l'image finale dans matplotlib (et elle peut contenir un ou plusieurs `Axes`), souvent abrégée en `fig`.
* Les **`Axes`** sont des tracés individuels (par exemple `ax = fig.add_subplot()`), souvent abrégés en `ax`.
    * Une `Figure` peut contenir un ou plusieurs `Axes`.
* Les **`Axes`** sont x (horizontal), y (vertical), z (profondeur).

Ajoutons maintenant quelques données à notre graphique précédent.

````python
# Ajouter des données à notre tracé précédent 
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
````

Mais il y a un moyen plus simple d'utiliser `matplotlib.pyplot` pour nous aider à créer une `Figure` avec de multiples `Axes` potentiels.

Et c'est avec [`plt.subplots()`] (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html).

````python
# Créer une figure et plusieurs axes potentiels et ajouter des données
fig, ax = plt.subplots()
ax.plot(x, y);
````

````python
# C'est de là que vient le nom "orienté objet". 
type(fig), type(ax)
````

### Un flux de travail rapide pour Matplotlib

Le flux de travail suivant est une pratique standard lors de la création d'un tracé matplotlib : 

0. **Importer matplotlib** - Par exemple, `import matplotlib.pyplot as plt`).
1. **Préparer les données** - Elles peuvent provenir d'un ensemble de données existant (analyse de données) ou des résultats d'un modèle d'apprentissage automatique (science des données).
2. **Setup the plot** - En d'autres termes, créez la figure et les différents axes.
3. **Tracer les données sur les axes** - Envoyez les données pertinentes sur les axes cibles.
4. **Cutomize the plot** - Ajoutez un titre, décorez les couleurs, étiquetez chaque axe.
5. **Enregistrer (facultatif) et montrer** - Voyez à quoi ressemble votre chef-d'œuvre et enregistrez-le dans un fichier si nécessaire.

````python
# Un flux de travail matplotlib

# 0. importer et préparer matplotlib
# %matplotlib inline # Pas nécessaire dans les nouvelles versions de Jupyter (par ex. à partir de 2022)
import matplotlib.pyplot as plt

# 1. Préparer les données
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Configurer le graphe (figure et axes)
fig, ax = plt.subplots(figsize=(10,10))

# 3. Tracer les données
ax.plot(x, y)

# 4. Personnaliser le tracé
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Sauvegarder et afficher
fig.savefig("../images/simple-plot.png")
````

## 2. Réaliser les types de tracés les plus courants à l'aide des tableaux NumPy

- Diagramme de dispersion - [`ax.scatter()`](https://matplotlib.org/stable/plot_types/basic/scatter_plot.html#sphx-glr-plot-types-basic-scatter-plot-py)

- Histogramme - [`ax.hist()`](https://matplotlib.org/stable/plot_types/stats/hist_plot.html#sphx-glr-plot-types-stats-hist-plot-py)

Nous allons voir comment toutes ces méthodes peuvent être créées à partir de [`matplotlob.pyplot.subplots()`] (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots).

**Ressource:** Rappelez-vous que vous pouvez [voir les différents types de tracés de matplotlib](https://matplotlib.org/stable/plot_types/index.html) dans la documentation.

````python
import numpy as np
````

### Création d'un tracé linéaire

Line est le type de visualisation par défaut dans Matplotlib. Habituellement, sauf indication contraire, vos tracés commenceront par des lignes.

Les tracés linéaires sont parfaits pour voir les tendances dans le temps.

````python
# Creer un tableau
x = np.linspace(0, 10, 100)
x[:10]
````

````python
# Le tracé par défaut est la ligne
fig, ax = plt.subplots()
ax.plot(x, x**2);
````

### Création d'un diagramme de dispersion

Les diagrammes de dispersion sont très utiles lorsque vous disposez de nombreux points de données différents et que vous souhaitez voir comment ils interagissent les uns avec les autres sans être reliés entre eux. 

````python
# Nécessité de recréer nos instances de figures et d'axes lorsque nous voulons une nouvelle figure
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
````

````python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
````

### Création de diagrammes à barres

Les diagrammes en bâtons sont très utiles pour visualiser différentes quantités d'articles à thème similaire.

Par exemple, les ventes d'articles dans un magasin de beurre de noix.

Vous pouvez créer des diagrammes à barres **verticaux** avec [`ax.bar()`](https://matplotlib.org/stable/plot_types/basic/bar.html#sphx-glr-plot-types-basic-bar-py) et des diagrammes à barres **horizontaux** avec [`ax.barh()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html).

````python
# Vous pouvez faire des tracés à partir d'un dictionnaire
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
````

````python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
````

### Création d'un histogramme

Les histogrammes sont excellents pour montrer la distribution des données.

Par exemple, vous pourriez vouloir montrer la distribution des âges d'une population ou des salaires d'une ville.

````python
# Fait des données à partir d'une distribution normale
x = np.random.randn(1000) # tire des données d'une distribution normale

fig, ax = plt.subplots()
ax.hist(x);
````

````python
x = np.random.random(1000) # données aléatoires provenant d'une distribution aléatoire

fig, ax = plt.subplots()
ax.hist(x);
````

### Création de figures avec plusieurs axes à l'aide de sous-graphes

Les sous-graphes vous permettent de créer plusieurs axes sur la même figure (plusieurs tracés à l'intérieur d'un même graphe).

Les sous-graphes sont utiles parce que vous commencez avec un seul graphe par figure, mais vous pouvez en créer d'autres si nécessaire.

Par exemple, créons un sous-graphe qui montre plusieurs des ensembles de données ci-dessus sur la même figure.

Nous pouvons le faire en créant plusieurs axes avec [`plt.subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots) et en réglant les paramètres `nrows` (nombre de lignes) et `ncols` (nombre de colonnes) pour refléter le nombre d'axes que nous souhaitons.

Les paramètres `nrows` et `ncols` sont multiplicatifs, ce qui signifie que `plt.subplots(nrows=2, ncols=2)` créera `2*2=4` axes au total. 

**Ressource:** Vous pouvez voir un [nombre sensationnel d'exemples] (https://matplotlib.org/stable/gallery/index.html#subplots-axes-and-figures) pour la création de Subplots dans la documentation de matplotlib. 

````python
# Option 1 : Créer 4 sous-graphes avec chaque axe ayant son propre nom de variable
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10, 5))

# Tracer les données sur chaque axe
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
````

````python
# Option 2 : Créer 4 sous-graphes avec une seule variable ax
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Indexer la variable ax pour tracer les données
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
````

### Tracé linéaire à partir d'un DataFrame pandas

Pour comprendre les exemples, je dois souvent les répéter (les coder moi-même) plutôt que de simplement les lire.

## 4. Tracer des graphiques plus avancés à partir d'un DataFrame pandas 


Il est possible d'obtenir des tracés beaucoup plus compliqués et détaillés à partir d'un DataFrame pandas.

Entraînons-nous à utiliser le DataFrame `heart_disease`. 

Et à titre d'exemple, effectuons une analyse sur les personnes âgées de plus de 50 ans.

Pour ce faire, commençons par créer un graphique directement à partir de pandas, puis utilisons l'API orientée objet (`plt.subplots()`) pour le développer.

Créons maintenant un diagramme de dispersion directement à partir du DataFrame pandas.

C'est assez facile à faire, mais c'est un peu limité en termes de personnalisation.

Visualisons les taux de cholestérol des patients de plus de 50 ans.

````python
# Créer un diagramme de dispersion directement à partir du DataFrame pandas
over_50.plot(kind="scatter",
             x="age", 
             y="chol", 
             c="target", # colorer les points en fonction de la valeur cible
             figsize=(10, 6));
````

Nous pouvons recréer le même tracé en utilisant `plt.subplots()` et en passant la variable Axes (`ax`) à la méthode pandas `plot()`.

````python
# Création d'une figure et d'une instance d'axes
fig, ax = plt.subplots(figsize=(10, 6))

# Tracer les données du DataFrame sur l'objet ax
over_50.plot(kind="scatter", 
             x="age", 
             y="chol", 
             c="target", 
             ax=ax); # définir les axes cibles

# Personnaliser les limites de l'axe des x (pour qu'elles correspondent à nos tranches d'âge cibles)
ax.set_xlim([45, 100]);
````

Maintenant, au lieu de tracer directement à partir du DataFrame pandas, nous pouvons faire un tracé un peu plus complet en traçant les données directement sur une instance d'Axes cible.

````python
# Création d'une figure et d'une instance d'axes
fig, ax = plt.subplots(figsize=(10, 6))

# Tracer les données directement sur les axes intance
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"]) # Colorer les données avec la colonne "cible".

# Personnaliser les paramètres de l'intrigue 
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Configurer la légende
ax.legend(*scatter.legend_elements(), 
          title="Target");
````

Et si nous voulions une ligne horizontale qui traverse la moyenne de `heart_disease["chol"]` ?

C'est ce que nous faisons avec la méthode [`Axes.axhline()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.axhline.html).

````python
# Créer le graphe
fig, ax = plt.subplots(figsize=(10, 6))

# Tracer les données
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])

# Personnaliser le tracé
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Ajouter une légende
ax.legend(*scatter.legend_elements(), 
          title="Target")

# Ajouter une ligne moyenne
ax.axhline(over_50["chol"].mean(),
           linestyle="--"); # style de la ligne pour qu'elle soit belle
````

### Tracer plusieurs tracés sur la même figure (ajouter un autre tracé à un tracé existant)

Il arrive que vous souhaitiez visualiser plusieurs caractéristiques d'un ensemble de données ou les résultats d'un modèle sur une même figure.

Vous pouvez y parvenir en ajoutant des données à plusieurs axes sur la même figure.

La méthode [`plt.subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html) vous aide à créer des Figures avec un nombre désiré d'Axes dans une configuration désirée.

En utilisant les paramètres `nrows` (nombre de lignes) et `ncols` (nombre de colonnes), vous pouvez contrôler le nombre d'axes sur la figure.

Par exemple :
* `nrows=2`, `ncols=1` = 2x1 = une Figure avec 2 Axes
* `nrows=5`, `ncols=5` = 5x5 = une figure avec 25 axes

Créons un graphique avec 2 axes.

Sur le premier axe (axe 0), nous représenterons les maladies cardiaques en fonction du taux de cholestérol (`chol`).

Sur le deuxième axe (axe 1), nous représenterons les maladies cardiaques en fonction des niveaux de fréquence cardiaque maximale (`thalach`).

````python
# Configuration du graphique (2 lignes, 1 colonne)
fig, (ax0, ax1) = plt.subplots(nrows=2, # 2 lignes
                               ncols=1, # 1 colonne 
                               sharex=True, # les deux graphiques doivent utiliser le même axe x 
                               figsize=(10, 8))

# ---------- Axe 0 : Maladies cardiaques et taux de cholestérol ----------

# Ajouter des données pour ax0
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"])
# Personnaliser ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")

# Établissement d'une ligne moyenne
ax0.axhline(y=over_50["chol"].mean(), 
            color='b', 
            linestyle='--', 
            label="Average")

# ---------- Axe 1 : Maladies cardiaques et niveaux maximaux de fréquence cardiaque ----------

# Ajouter les données pour ax1
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"])

# Personnaliser ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")

# Établissement d'une ligne moyenne
ax1.axhline(y=over_50["thalach"].mean(), 
            color='b', 
            linestyle='--', 
            label="Average")

# Titre de la figure
fig.suptitle('Heart Disease Analysis', 
             fontsize=16, 
             fontweight='bold');
````

### Personnaliser les couleurs des graphiques avec les cartes de couleurs (cmap)

La couleur est l'une des caractéristiques les plus importantes d'un graphique.

Elle permet de distinguer différents types d'informations.

Et avec les bonnes couleurs, les graphiques peuvent être amusants à regarder et à essayer d'en apprendre plus.

Matplotlib fournit de nombreuses options de couleurs grâce à [`matplotlib.colormaps`] (https://matplotlib.org/stable/api/matplotlib_configuration_api.html#matplotlib.colormaps).

Voyons comment nous pouvons changer les couleurs d'un graphique Matplotlib via le paramètre `cmap` (`cmap` est l'abréviation de `colormaps`).

Nous allons commencer par créer un diagramme de dispersion avec la valeur `cmap` par défaut (`cmap="viridis"`).

````python
# Configuration de la figure et des axes
fig, ax = plt.subplots(figsize=(10, 6))

# Créer un diagramme de dispersion sans changement de carte (utiliser la carte des couleurs par défaut)
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"],
                     cmap="viridis") # valeur par défaut de la cmap

# Ajouter des attributs au tracé
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
ax.axhline(y=over_50["chol"].mean(), 
           c='b', 
           linestyle='--', 
           label="Average");
ax.legend(*scatter.legend_elements(), 
          title="Target");
````

Merveilleux !

Cette parcelle n'a pas l'air trop mal.

Mais que se passe-t-il si nous voulons changer les couleurs ?

Il existe de nombreuses options de paramètres `cmap` disponibles dans la [référence de la carte des couleurs] (https://matplotlib.org/stable/gallery/color/colormap_reference.html).

Et si nous essayions `cmap="winter"` ? 

Nous pouvons également changer la couleur de la ligne horizontale en utilisant le paramètre `color` et en lui attribuant une chaîne de caractères de la couleur souhaitée (par exemple, `color="r"` pour rouge).

````python
fig, ax = plt.subplots(figsize=(10, 6))

# Configurer le diagramme de dispersion avec différentes cmap
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"], 
                     cmap="winter") # Changer la valeur de cmap

# Ajouter des attributs au tracé avec une ligne de couleur différente
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")
ax.axhline(y=over_50["chol"].mean(), 
           color="r", # Changer la couleur de la ligne en "r" (pour rouge)
           linestyle='--', 
           label="Average");
ax.legend(*scatter.legend_elements(), 
          title="Target");
````

Woohoo !

Le premier graphique était joli, mais je pense que je préfère les couleurs de ce nouveau graphique.

Pour en savoir plus sur [choosing colormaps in matplotlib] (https://matplotlib.org/stable/tutorials/colors/colormaps.html), il existe un tutoriel sensationnel et approfondi dans la documentation de matplotlib. 

### Personnaliser les xlim et ylim

Matplotlib est assez efficace pour définir les plages de valeurs sur l'axe des x et l'axe des y.

Mais comme vous l'avez deviné, vous pouvez les personnaliser pour répondre à vos besoins.

Vous pouvez modifier les plages de valeurs des différents axes en utilisant les paramètres `xlim` et `ylim` à l'intérieur de la méthode `set()`.

Pour nous entraîner, recréons notre graphique à deux axes avec les valeurs par défaut des axes x et y. Nous ajouterons les mises à jour des couleurs dans le graphique à deux axes.

Nous ajouterons également les mises à jour des couleurs de la section précédente.

````python
# Recréer le graphique à deux axes ci-dessus avec des mises à jour des couleurs 
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               sharex=True, 
                               figsize=(10, 7))

# ---------- Axe 0 ----------
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap="winter")
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

# Établir une ligne de moyenne
ax0.axhline(y=over_50["chol"].mean(), 
            color="r", 
            linestyle="--", 
            label="Average");
ax0.legend(*scatter.legend_elements(), title="Target")

# ---------- Axe 1 ----------
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap="winter")
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Établir une ligne de moyenne
ax1.axhline(y=over_50["thalach"].mean(), 
            color="r", 
            linestyle="--", 
            label="Average");
ax1.legend(*scatter.legend_elements(), 
           title="Target")

# Titre de la figure
fig.suptitle("Heart Disease Analysis", 
             fontsize=16, 
             fontweight="bold");
````

Recréons maintenant le tracé ci-dessus, mais cette fois-ci, nous allons changer les limites des axes.

Nous pouvons le faire en utilisant [`Axes.set(xlim=[50, 80])`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set.html) ou `Axes.set(ylim=[60, 220])` où les entrées de `xlim` et `ylim` sont une liste d'entiers définissant une plage de valeurs.

Par exemple, `xlim=[50, 80]` fixera les valeurs de l'axe des x pour commencer à `50` et finir à `80`.

````python
# Recréer le graphique ci-dessus avec des plages d'axes x et y personnalisées
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1, 
                               sharex=True, 
                               figsize=(10, 7))
scatter = ax0.scatter(over_50["age"], 
                      over_50["chol"], 
                      c=over_50["target"],
                      cmap='winter')
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol",
        xlim=[50, 80]) # définir les plages de l'axe des x 

# Établir une ligne de moyenne
ax0.axhline(y=over_50["chol"].mean(), 
            color="r", 
            linestyle="--", 
            label="Average");
ax0.legend(*scatter.legend_elements(), title="Target")

# Axe 1, 1 (ligne 1, colonne 1)
scatter = ax1.scatter(over_50["age"], 
                      over_50["thalach"], 
                      c=over_50["target"],
                      cmap='winter')
ax1.set(title="Heart Disease and Max Heart Rate Levels",
        xlabel="Age",
        ylabel="Max Heart Rate",
        ylim=[60, 220]) # changer la plage de l'axe des y

# Établir une ligne de moyenne
ax1.axhline(y=over_50["thalach"].mean(), 
            color="r", 
            linestyle="--", 
            label="Average");
ax1.legend(*scatter.legend_elements(), 
           title="Target")

# Titre de la figure
fig.suptitle("Heart Disease Analysis", 
             fontsize=16, 
             fontweight="bold");
````
